```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem, Draw
import random

# Reproducibility
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

# Configuration
N_BITS = 1024  # Morgan Fingerprint size
INPUT_DIM = N_BITS * 2  # Two drugs concatenated
print("Libraries loaded. System ready.")
```

```python
def generate_data():
    # Valid SMILES pool (Real chemical structures)
    valid_smiles = [
        "CC(=O)OC1=CC=CC=C1C(=O)O", # Aspirin
        "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O", # Ibuprofen
        "CC(=O)NC1=CC=C(O)C=C1", # Paracetamol
        "CN1C=NC2=C1C(=O)N(C(=O)N2C)C", # Caffeine
        "CC(=O)C(C1=CC=CC=C1)C2=C(O)C3=CC=CC=C3OC2=O", # Warfarin
        "C1=CC=C(C=C1)C=O", # Benzaldehyde
    ]
    
    drug_names = ["Aspirin", "Ibuprofen", "Paracetamol", "Caffeine", "Warfarin", "Benzaldehyde"]
    drug_db = {name: s for name, s in zip(drug_names, valid_smiles)}
    
    # Generate random interaction pairs
    interactions = []
    keys = list(drug_db.keys())
    
    for _ in range(200):
        d1 = random.choice(keys)
        d2 = random.choice(keys)
        if d1 == d2: continue
        
        # Random label (0=Safe, 1=interaction)
        label = np.random.randint(0, 2)
        interactions.append({
            "Drug_A": d1, "Drug_B": d2,
            "SMILES_A": drug_db[d1], "SMILES_B": drug_db[d2],
            "Label": label
        })
        
    return drug_db, pd.DataFrame(interactions)

drug_db, df = generate_data()
print(f"Generated {len(df)} interaction pairs.")
df.head()
```

```python
# Visualize the first 4 unique drugs in our database
unique_smiles = list(drug_db.values())[:4]
unique_names = list(drug_db.keys())[:4]

mols = [Chem.MolFromSmiles(s) for s in unique_smiles]
img = Draw.MolsToGridImage(
    mols, 
    molsPerRow=4, 
    subImgSize=(200,200), 
    legends=unique_names
)
display(img)
```

```python
from rdkit.Chem import rdFingerprintGenerator

# 1. Initialize the Fingerprint Generator once (Best Practice)
# We use the new Generator API instead of the old 'AllChem' calls
morgan_gen = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=N_BITS)

def smiles_to_vector(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None: 
        return np.zeros(N_BITS)
    
    # 2. Generate Fingerprint (No warnings)
    fp = morgan_gen.GetFingerprint(mol)
    
    # 3. Convert to NumPy
    # We can simply pass the RDKit object to np.array now
    return np.array(fp)

# Prepare Training Data
X_data = []
y_data = []

for _, row in df.iterrows():
    v1 = smiles_to_vector(row['SMILES_A'])
    v2 = smiles_to_vector(row['SMILES_B'])
    
    # Concatenate vectors
    combined = np.concatenate([v1, v2])
    
    X_data.append(combined)
    y_data.append(row['Label'])

# Convert to PyTorch Tensors
X_tensor = torch.FloatTensor(np.array(X_data))
y_tensor = torch.FloatTensor(np.array(y_data)).unsqueeze(1)

print(f"Feature Matrix Shape: {X_tensor.shape}")
print("✅ Vectorization complete without warnings.")
```

```python
class DDI_Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(INPUT_DIM, 512)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.3)
        self.layer2 = nn.Linear(512, 128)
        self.output = nn.Linear(128, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        x = self.dropout(self.relu(self.layer1(x)))
        x = self.relu(self.layer2(x))
        return self.sigmoid(self.output(x))

model = DDI_Net()
print(model)
```

```python
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.BCELoss()

epochs = 50
losses = []

model.train()
for epoch in range(epochs):
    optimizer.zero_grad()
    pred = model(X_tensor)
    loss = criterion(pred, y_tensor)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())
    
    if epoch % 10 == 0:
        print(f"Epoch {epoch} | Loss: {loss.item():.4f}")

print("Training Complete.")
```

```python
# Test Case: Aspirin + Warfarin
drug_a = "Aspirin"
drug_b = "Warfarin"

vec_a = smiles_to_vector(drug_db[drug_a])
vec_b = smiles_to_vector(drug_db[drug_b])
input_vec = torch.FloatTensor(np.concatenate([vec_a, vec_b]))

model.eval()
with torch.no_grad():
    prob = model(input_vec).item()

print(f"Interaction Probability between {drug_a} and {drug_b}: {prob:.4f}")
```
